## **G. Probabilities**

### 1. Compute the probability that randomly selected country with Happiness score over 6.0 is from Western Europe. You will have to use pandas to count the appropriate quantities.

```python
import numpy as np
import random


index = df.index # there are 2 indexes in the dataframe...the ordered one is after grouping by Happiness Score per region
number_of_rows = len(index) # get the total number of rows

experiments = 100000
counter = 0
prob = 0


for i in range(experiments):
    cnt = random.randrange(number_of_rows) # select randomly one of the 151 countries
    
    df_row = df.iloc[cnt] # loc the specific row
    
    if ((df_row['Region']== 'Western-Europe') & (df_row['Happiness_Score']>6.0)):
        counter += 1

    prob = counter / experiments

print("The probability that a randomly selected country with hapiness score over 6.0 is from WE, is: {:.2f}%".format(prob*100))
```
